```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("/Users/namirsacic/tokenized-assets/data_analysis/processed_data/daily_merged_data/AAPL_daily_merged.csv")
# Assume `df` is your dataframe with columns 'date' and 'volume'
df['day'] = pd.to_datetime(df['day']) # convert date column to datetime
df.set_index('day', inplace=True) # set date column as index

# Resample the data to get daily volume
daily_volume = df.resample('D').sum()

# Calculate the weekly moving average
weekly_ma = daily_volume.rolling(window=7, center=True).mean()

# Create the bar chart with moving average line
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(daily_volume.index, daily_volume['size'], width=1.0, label='Daily Volume')
ax.plot(weekly_ma.index, weekly_ma['size'], color='red', linewidth=2, label='Weekly Moving Average')
ax.set_xlabel('Day')
ax.set_ylabel('Trading Size')
ax.set_title('Traded Volume per Day with Weekly Moving Average')
ax.legend() # Add the legend to the plot
plt.show()





```

```python
# Create the histogram
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(df['premium'], bins=100, alpha=0.7)

# Customize the appearance of the plot
ax.set_ylabel('Count', fontsize=14)
ax.set_xlabel('Time', fontsize=14)
ax.set_title('Distribution of Premium over Time', fontsize=18)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(['Premium'], loc='upper right', fontsize='large')

# Invert the y-axis
ax.invert_yaxis()

# Show the plot
plt.show()
```
