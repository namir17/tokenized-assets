```python
def df_preprocess(filename):
    
    pathname = '/Users/sophiawang/Desktop/Frankfurt School/Guided Study Financial Management/Group Project/trades'
    
    df = pd.read_csv(os.path.join(pathname, filename))

    processed_df = pd.DataFrame(columns = ['Average Price', 'Average Size', 'Weighted Average Price', 'Absolute Volume', 
                                           'Relative Volume', 'Count'])

    df['time'] = pd.to_datetime(df['time'], errors='coerce')
    df['date'] = df['time'].dt.date

    processed_df['Average Price'] = df.groupby('date').agg({'price': 'mean'})
    processed_df['Average Size'] = df.groupby('date').agg({'size': 'mean'})
    processed_df['Weighted Average Price'] = df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())
    processed_df['Absolute Volume'] = df.groupby('date').apply(lambda x: (x['price']*x['size']).sum())
    #processed_df['Accumulative Volume'] 
    df['side'] = df['side'].replace({'buy': 1, 'sell': -1})
    df['Adjusted_size'] = df['side'] * df['size']
    processed_df['Relative Volume'] = df.groupby('date')['Adjusted_size'].sum()
    processed_df['Relative Volume'] = processed_df['Relative Volume'].cumsum()
    processed_df['Count'] = df.groupby('date')['id'].count()
    buy_df = df[df['side']== 1]
    sell_df = df[df['side']== -1]

    processed_df['WAP_Buy'] = buy_df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())
    processed_df['WAP_Sell'] = sell_df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())

    return processed_df
```
