```python
pathname = '/Users/namirsacic/tokenized-assets/ftx/'

# Define a dictionary to store the dataframes for each class
class_dict = {
    'Large': ['AAPLUSD', 'ACBUSD', 'AMDUSD', 'AMZNUSD', 'APHAUSD', 'BABAUSD', 'BILIUSD', 'BNTXUSD', 'BYNDUSD', 'FBUSD', 'GBTCUSD', 'GLXYUSD', 'GOOGLUSD', 'MRNAUSD', 'MSTRUSD', 'NFLXUSD', 'NOKUSD', 'NVDAUSD', 'PYPLUSD', 'SQUSD', 'TSLABTC', 'TSLAUSD', 'TSMUSD', 'TWTRUSD', 'UBERUSD', 'ZMUSD'],
    'Meme': ['AMCUSD', 'ARKKUSD', 'CGCUSD', 'GMEUSD', 'HOODUSD'],
    'Crypto': ['BITOUSD', 'BITWUSD', 'ETHEUSD', 'GBTCUSD'],
    'Commodity': ['GDXJUSD', 'GDXUSD', 'GLDUSD', 'USOUSD'],
    'Other': ['ABNBUSD', 'BBUSD', 'CRONUSD', 'DKNGUSD', 'TSLADOGE', 'TLRYUSD', 'WNDRUSD'],
    'SP500':['SPYUSD']
}

# Create an empty list to store the dataframes
df_lists = []

# Loop through the dictionary and read in the csv files for each class
for key, value in class_dict.items():
    for ticker in value:
        df = pd.read_csv(os.path.join(pathname, f"ftx_{ticker}.csv"))
        df['ticker'] = ticker[:4]
        df['class'] = key
        df['Volume'] = df['price'] * df['size']
        df_lists.append(df)

# Concatenate all the dataframes into one
combined_ftx_raw_df = pd.concat(df_lists)
```

```python
def df_preprocess(df):
    
    processed_df = pd.DataFrame(columns = ['Average Price', 'Average Size', 'Weighted Average Price', 'Absolute Volume', 
                                           'Relative Volume', 'Count'])

    df['time'] = pd.to_datetime(df['time'], errors='coerce')
    df['date'] = df['time'].dt.date

    processed_df['Average_Price'] = df.groupby('date').agg({'price': 'mean'})
    processed_df['Average_Size'] = df.groupby('date').agg({'size': 'mean'})
    processed_df['Weighted_Average_Price'] = df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())
    processed_df['Absolute_Volume'] = df.groupby('date').apply(lambda x: (x['price']*x['size']).sum())
    #processed_df['Accumulative Volume'] 
    df['side'] = df['side'].replace({'buy': 1, 'sell': -1})
    df['Adjusted_size'] = df['side'] * df['size']
    processed_df['Relative_Volume'] = df.groupby('date')['Adjusted_size'].sum()
    processed_df['Relative_Volume'] = processed_df['Relative_Volume'].cumsum()
    processed_df['Count'] = df.groupby('date')['id'].count()
    buy_df = df[df['side']== 1]
    sell_df = df[df['side']== -1]

    processed_df['WAP_Buy'] = buy_df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())
    processed_df['WAP_Sell'] = sell_df.groupby('date').apply(lambda x: (x['price']*x['size']).sum() / x['size'].sum())

    return processed_df
```

```python
daily_folder = '/Users/namirsacic/tokenized-assets/daily_data'

# list all files in the daily folder
files = os.listdir(daily_folder)

daily_files_df = []
for file in files:
    df = pd.read_csv(os.path.join(daily_folder, file))
    ticker = file.split(".")[0]
    if ticker == "BITO":
        df["class"] = "Crypto"
    elif ticker in ["GME", "HOOD", "AMC", "ARKK"]:
        df["class"] = "Meme"
    elif ticker in ["GLD", "GDX", "GDXJ", "SLV", "USO"]:
        df["class"] = "Commodity"
    elif ticker == "SPY":
        df["class"] = "SP500"
    elif ticker in ["ACB", "WEED", "DKNG", "BYND", "CRON", "TLRY", "GLXY", "BB"]:
        df["class"] = "other"
    else:
        df["class"] = "Large"
    df["ticker"] = ticker
    daily_files_df.append(df)
```

```python
com_daily_underlying = pd.concat(daily_files_df)
com_daily_underlying['Average_HCLO_Price'] = (com_daily_underlying['HIGH'] + com_daily_underlying['CLOSE'] + com_daily_underlying['LOW'] + com_daily_underlying['OPEN'])/4
com_daily_underlying['Daily_Total_Absolute_Volume'] = com_daily_underlying['Average_HCLO_Price'] * com_daily_underlying['VOLUME']

com_daily_underlying_large = com_daily_underlying[com_daily_underlying['class']=='Large']
com_daily_underlying_meme = com_daily_underlying[com_daily_underlying['class']=='Meme']
com_daily_underlying_crypto = com_daily_underlying[com_daily_underlying['class']=='Crypto']
com_daily_underlying_commodity = com_daily_underlying[com_daily_underlying['class']=='Commodity']
com_daily_underlying_other = com_daily_underlying[com_daily_underlying['class']=='other']
com_daily_underlying_SP500 = com_daily_underlying[com_daily_underlying['class']=='SP500']
```

```python
def transfer_to_weekly(daily_df):
    daily_df['Date'] = pd.to_datetime(daily_df['Date'])
    weekly_df = daily_df.groupby(pd.Grouper(key='Date', freq='W'))['Daily_Total_Absolute_Volume'].sum()
    weekly_df = pd.DataFrame(weekly_df).reset_index()
    weekly_df = weekly_df.rename(columns={'Date': 'Weekly_Date', 'Daily_Total_Absolute_Volume': 'Weekly_Total_Absolute_Volume'})
    return weekly_df
```

```python
com_weekly_underlying_large = transfer_to_weekly(com_daily_underlying_large)
com_weekly_underlying_meme = transfer_to_weekly(com_daily_underlying_meme)
com_weekly_underlying_crypto = transfer_to_weekly(com_daily_underlying_crypto)
com_weekly_underlying_commodity = transfer_to_weekly(com_daily_underlying_commodity)
com_weekly_underlying_other = transfer_to_weekly(com_daily_underlying_other)
com_weekly_underlying_SP500 = transfer_to_weekly(com_daily_underlying_SP500)
```
