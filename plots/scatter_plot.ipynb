```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define the list of ticker symbols
tickers = ['AAPL', 'ABNB', 'ACB', 'AMC', 'AMZN', 'ARKK', 'BABA', 'BB', 'BITO',
           'BNTX', 'BYND', 'COIN', 'CRON', 'DKNG', 'GDXJ', 'GDX', 'GLD',
           'GME', 'GOOGL', 'HOOD', 'MRNA', 'MSTR', 'NFLX', 'NIO', 'NVDA', 'PENN', 'PFE', 'PYPL', 'SLV',
           'SPY', 'SQ', 'TLRY', 'TSLA', 'TSLA', 'TSLA', 'TWTR', 'UBER', 'USO', 'ZM']
```

```python
pathname = '/Users/namirsacic/tokenized-assets/data_analysis/processed_data/merged_data'

# Create an empty list to store the dataframes
df_lists = []

# Loop over the list of tickers and create a new dataframe for each ticker
for ticker in tickers:
    filename = f'{ticker}_merged.csv'
    filepath = os.path.join(pathname, filename)
    df = pd.read_csv(filepath)
    df_lists.append(df)
```

```python
# Concatenate the list of dataframes into a single dataframe
combined_df = pd.concat(df_lists)
```

```python
# Set the size of the figure
fig, ax = plt.subplots(figsize=(12, 8))

# Create the scatter plot
ax.scatter(combined_df['price*size'], combined_df['premium'], alpha=0.5, c='steelblue')

# Add axis labels and a title
ax.set_xlabel('Trade Size')
ax.set_ylabel('Price Quality')
ax.set_title('Trade Size vs. Price Quality Scatter Plot')

# Add a grid
ax.grid(True)

# Add some fancy styles
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the plot
plt.show()
```
