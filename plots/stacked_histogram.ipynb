```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the folder path where your CSV files are stored
folder_path = '/Users/namirsacic/tokenized-assets/data_analysis/processed_data/daily_merged_data/'

# Get a list of all CSV files in the folder
csv_files = glob.glob(folder_path + '*.csv')

# Create an empty list to store the data frames
dfs = []

# Loop through all CSV files and load them into data frames
for file in csv_files:
    df = pd.read_csv(file)
    df['day'] = pd.to_datetime(df['day']) # Convert the 'day' column to datetime format
    dfs.append(df) # Add the data frame to the list

# Set the percentile range to scale the histogram
percentile_range = [0.01, 99.9]

# Calculate the limits of the x-axis based on the percentile range
x_min, x_max = pd.concat(dfs)['premium'].quantile(percentile_range[0]/100), pd.concat(dfs)['premium'].quantile(percentile_range[1]/100)

# Define a monochromatic blue color palette for the histograms
colors = sns.color_palette('Blues', len(dfs))

# Create stacked histograms of the 'premium' column with time on the x-axis
sns.set_style('darkgrid')
plt.hist([df['premium'] for df in dfs], bins=50, range=(x_min, x_max), stacked=True, label=[df.columns[0].split('_')[0] for df in dfs], color=colors, alpha=0.8)
plt.xlabel('Premium')
plt.ylabel('Frequency')
plt.yscale("log")
plt.title('Premium Histogram', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#plt.legend(fontsize=12)
sns.despine()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV files
df = pd.read_csv('AAPL_daily_merged.csv')
df1 = pd.read_csv('ABNB_daily_merged.csv')
df2 = pd.read_csv('AMC_daily_merged.csv')

# Convert the 'time' column to datetime format
df['day'] = pd.to_datetime(df['day'])
df1['day'] = pd.to_datetime(df1['day'])
df2['day'] = pd.to_datetime(df2['day'])

# Set the percentile range to scale the histogram
percentile_range = [1, 95]

# Calculate the limits of the x-axis based on the percentile range
x_min, x_max = df['premium'].quantile(percentile_range[0]/100), df['premium'].quantile(percentile_range[1]/100)

# Define a list of colors for the datasets
colors = ['steelblue', 'red', 'green', 'orange', 'purple', 'yellow', 'brown', 'grey', 'pink', 'olive', 'teal', 'navy', 'maroon', 'tan', 'lavender', 'cyan', 'magenta', 'gold', 'salmon', 'orchid', 'turquoise', 'plum', 'beige', 'azure', 'indigo', 'bisque', 'coral', 'crimson', 'fuchsia', 'honeydew', 'lime', 'peru', 'sienna', 'violet', 'wheat', 'crimson', 'darkorange']

# Create stacked histograms of the 'premium' column with time on the x-axis
sns.set_style('whitegrid')
plt.hist([df['premium'], df1['premium'], df2['premium']], bins=50, range=(x_min, x_max), stacked=True, label=['AAPL', 'ABNB', 'AMC'], color=colors[:3], alpha=0.8)
plt.xlabel('Premium')
plt.ylabel('Frequency')
plt.title('Premium Histogram', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
sns.despine()
plt.show()
```
